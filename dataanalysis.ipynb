# Chicago Car Crashes

## Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load The Dataset

```python
df = pd.read_csv("data/Traffic_Crashes_-_Crashes.csv")
```

## Clean the Data

```python
df
df = df.drop_duplicates()
df = df[['CRASH_DATE', 'POSTED_SPEED_LIMIT','ROAD_DEFECT','DAMAGE','INJURIES_TOTAL','INJURIES_FATAL','CRASH_HOUR', 'CRASH_DAY_OF_WEEK', 'CRASH_MONTH']]
#remove rows where the speed limit is imposibly high
df = df.loc[df['POSTED_SPEED_LIMIT'] < 90]
df
```

## Univariate Analysis

```python
sns.histplot(data= df, x= "POSTED_SPEED_LIMIT", binwidth= 5)
```

```python
sns.histplot(data= df, x= "INJURIES_TOTAL", binwidth= 1)
```

```python
sns.histplot(data= df, x= "CRASH_MONTH", binwidth= 1)
```

```python
sns.histplot(data= df, x= "DAMAGE")
```

```python
sns.scatterplot(data= df, x= 'POSTED_SPEED_LIMIT', y = 'INJURIES_TOTAL')
plt.title("Speed vs Num Injuries")
```

## What's the most dangerous time of day to be driving?

```python
sns.scatterplot(data= df, x= 'CRASH_HOUR', y = 'INJURIES_TOTAL')
plt.title("Time of Day vs Injuries")
```

```python

sns.scatterplot(data= df, x= 'CRASH_HOUR', y = 'INJURIES_FATAL')
plt.title("Time of Day vs Fatal Injuries")
```

Fram this we can see that driving at any time during the day is likely to cause more injuries, however the risk of fatal injuries increases at night when the total number of injuries are lower. 

## Does a higher speed result in a more expensive repair?

```python
sns.scatterplot(data= df, x= 'POSTED_SPEED_LIMIT', y = 'DAMAGE')
plt.title("Posted Speed Limit vs Damage Cost")
```

We can see that there is a bit of a skew towards more expensive damage with higher speed limits, however it isn't quite as extreme as you'd think, and this may partially be a result of the fact that damage expenses are lumped into only 3 categories.

## How does the weekend change when crashes are most likely to happen?

```python
sns.scatterplot(data= df, x= 'CRASH_DAY_OF_WEEK', y = 'POSTED_SPEED_LIMIT')
plt.title("Day of Week vs Posted Speed Limit")
```

```python
sns.histplot(data= df, x= 'CRASH_DAY_OF_WEEK')
```

Saturday is the peak of the number of crashes, and the weekend is a bit higher than the rest of the week. However the number of crashes dont super dramatically increase on the weekends.
